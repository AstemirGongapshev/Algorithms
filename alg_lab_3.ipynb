3.1 Необходимо реализовать очередь c приоритетом со всеми методами 

```python
class PriorityQueue:
    def __init__(self, *args):
        self.queue = []
        if args:
            for item in args:
                self.enqueue(*item)

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item, priority):
        self.queue.append((item, priority))
        self.queue.sort(key=lambda x: x[1], reverse=True)

    def dequeue(self):
        if self.is_empty():
            raise Exception("Очередь пуста")
        return self.queue.pop(0)[0]

    def peek(self):
        if self.is_empty():
            raise Exception("Очередь пуста")
        return self.queue[0][0]

    def empty(self):
        self.queue = []

    def size(self):
        return len(self.queue)

    def __str__(self):
        return '\n'.join(f'object: {item[0]}, priority: {item[1]}' for item in self.queue)

    def __getitem__(self, index):
        if not isinstance(index, int):
            raise TypeError(f'Index must be integer, but {index} was given')
        if -len(self.queue) <= index < len(self.queue):
            return self.queue[index]
        else:
            raise IndexError(f'Index {index} is out of range')

    def __iter__(self):
        self.iter = 0
        return self

    def __next__(self):
        if self.iter < len(self.queue):
            item = self.queue[self.iter]
            self.iter += 1
            return item
        else:
            raise StopIteration
        


# Провекра:
 # 1 Добавление
               
pq = PriorityQueue()
pq.enqueue('акция 1', 32)
pq.enqueue('акция 2', 60)
pq.enqueue('акция 4', 2)
pq.enqueue('акция 10', 11)
pq.enqueue('акция 24', 15)
pq.enqueue('акция 12', 12)
pq.enqueue('акция 101', 1)

print("Очередь после добавления элементов:")
print(pq)

# Удаление

removed_item = pq.dequeue()
print("Удалённый элемент:", removed_item)
print("Очередь после удаления:")
print(pq)


print("Очередь пуста?", pq.is_empty())
print("Размер очереди:", pq.size())


print("Итерация по элементам очереди:")
for item in pq:
    print(item)

peeked_item = pq.peek()
print("Первый элемент в очереди:", peeked_item)


pq.empty()
print("Очередь после очистки:")
print(pq)


pq.empty()  
try:
    pq.dequeue()
except Exception as e:
    print("Ошибка:", e)

try:
    pq.peek()
except Exception as e:
    print("Ошибка:", e)
```

3.2Необходимо найти максимальное произведение соседей в массиве, если несколько одинаковых результатов, то не обходимо выбрать с наименьшим множителем

```python
def find_max_(arr):
    if len(arr) < 2:
        return "Недостаточно элементов"

    max_product = arr[0] * arr[1]
    min_factor = min(arr[0], arr[1])
    pair = (arr[0], arr[1])

    for i in range(1, len(arr) - 1):
        product = arr[i] * arr[i + 1]
        current_min_factor = min(arr[i], arr[i + 1])

        if product > max_product or (product == max_product and current_min_factor < min_factor):
            max_product = product
            min_factor = current_min_factor
            pair = (arr[i], arr[i + 1])

    return pair, max_product


arr = [10, 4, 3, 5, 8, 0]
arr1 = [22,33,1,2,4,5,6,2,]
arr3 = [5,6,3,1,2,3,4,5,67,8]

result = find_max_(arr)
result1 = find_max_(arr1)
result3 = find_max_(arr3)

print("Пара с максимальным произведением:", result[0], "Произведение:", result[1])
print("Пара с максимальным произведением:", result1[0], "Произведение:", result1[1])
print("Пара с максимальным произведением:", result3[0], "Произведение:", result3[1])
```
